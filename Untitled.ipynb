```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

filepath = r'E:\data\gdp_1960_2020.csv'
df = pd.read_csv(filepath)

# Print the column names
print(df.columns)
```

# Bar chart

```python
# Define function to filter data by specified countries
def filter_data(df, countries):
    # Filter the data by the specified countries
    filtered_df = df[df['country'].isin(countries)]
    return filtered_df

# Define function to plot filtered data
def plot_data(filtered_df):
    # Create a bar plot of the GDP data for the specified countries
    plt.bar(filtered_df['country'], filtered_df['gdp'])
    plt.title('GDP by Country')
    plt.xlabel('Country')
    plt.ylabel('GDP (trillions of US dollars)')
    plt.xticks(rotation=90)
    plt.show()

# Specify the file path and countries to plot
file_path = 'E:/data/gdp_1960_2020.csv'
countries = ['United Kingdom', 'United States', 'Brazil', 'France', 'India', 'Israel']

# Read the data from CSV file
df = pd.read_csv(file_path)

# Filter the data by countries and plot the filtered data
filtered_df = filter_data(df, countries)
plot_data(filtered_df)
```

# Pie chart

```python
# Define a function to create a pie chart of GDP by country for specific countries
def pie_chart_countries(dataframe, countries):
    # Filter the data for the specific countries
    filtered_df = dataframe[dataframe['country'].isin(countries)]
    
    # Group the data by country and calculate the total GDP
    country_gdp = filtered_df.groupby('country')['gdp_percent'].sum()
    
    # Create a pie chart of the GDP by country
    plt.pie(country_gdp, labels=country_gdp.index, autopct='%1.1f%%')
    plt.title('GDP by Country')
    
    # Show the plot
    plt.show()

# Call the function with the specific countries and DataFrame
pie_chart_countries(df, ['United Kingdom', 'United States', 'Brazil', 'France', 'India', 'Israel'])
```

# Line plot

```python
# Select the countries to plot
countries = ['United Kingdom', 'United States', 'Brazil', 'France', 'India', 'Israel']

# Create a new figure
fig, ax = plt.subplots()

# Loop through the countries and plot the data
for country in countries:
    # Filter the DataFrame to include only the selected country
    df_country = df[df['country'] == country]
    
    # Get the year and GDP data
    years = df_country['year']
    gdp = df_country['gdp']
    
    # Generate sine and cosine functions to modify the data
    sine_data = np.sin(years) * gdp.max() / 10
    cosine_data = np.cos(years) * gdp.max() / 10
    
    # Create a line plot of modified GDP vs. year for the country
    ax.plot(years, gdp + sine_data - cosine_data, label=country)

# Set title and axis labels
plt.title('GDP by Year for Selected Countries')
plt.xlabel('Year')
plt.ylabel('GDP (trillions of US dollars)')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
